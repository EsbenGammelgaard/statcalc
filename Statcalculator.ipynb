```python
import numpy as np
from tkinter import *

root = Tk()
root.title('Statcalculator 2.0')
root.geometry("400x150")
root.columnconfigure(0,weight = 1)
root.columnconfigure(1,weight = 3)
root.columnconfigure(2,weight = 3)
root.columnconfigure(3,weight = 1)

def myClick():
    uds1 = int(uds11.get())
    resp1 = int(resp11.get())
    uds2 = int(uds22.get())
    resp2 = int(resp22.get())

    hyppighedkontrol = resp1 / uds1
    hyppighedtest = resp2 / uds2
    samlethyppighed = (resp1 + resp2) / (uds1 + uds2)
    varians = (1 / uds1 + 1 / uds2) * samlethyppighed * (1 - samlethyppighed)
    teststørrelse = (hyppighedkontrol - hyppighedtest) / np.sqrt(varians)

    if teststørrelse < 0 :
        teststørrelse = teststørrelse * -1

    if resp1 > uds1 or resp2 > uds2 :
        a = "Fejl! Din respons er større end din udsendelsesgruppe. Ret tallet"

    if teststørrelse <= 1.282 : 
        a = "Der er ingen signifikant forskel af test- og kontrolgruppen"

    if 1.282 < teststørrelse and teststørrelse < 1.695 :
        a = "Der er med 80% sandsynlighed forskel på test- og kontrolgruppe"

    if 1.695 <= teststørrelse and teststørrelse < 1.96 :
        a = "Der er med 90% sikkerhed forskel på test- og kontrolgruppe"

    if 2.576 > teststørrelse and teststørrelse > 1.96 :
        a = "Der er med 95% sikkerhed forskel på test- og kontrolgruppen"

    if teststørrelse >= 2.576 :
        a = "Der er med mere end 99% sikkerhed forskel på test- og kontrolgruppe"
        
    mylabel = Label(root,text=a).grid(row=5,column=0,columnspan=3)


mylabel1 = Label(root,text="Udsendelse stk").grid(row=0,column=1)
mylabel2 = Label(root,text="Svar stk").grid(row=0,column=2)
mylabel3 = Label(root,text="Kontrolgruppe").grid(row=1,column=0)
mylabel4 = Label(root,text="Testgruppe").grid(row=2,column=0)


uds11 = Entry(root, width=20)
uds11.grid(row=1,column=1)
resp11 = Entry(root, width=20)
resp11.grid(row=1,column=2)
uds22 = Entry(root, width=20)
uds22.grid(row=2,column=1)
resp22 = Entry(root, width=20)
resp22.grid(row=2,column=2)


myButton = Button(root, text="Udregn", command = myClick).grid(row=4,column=0,columnspan=3)



root.mainloop()
```
